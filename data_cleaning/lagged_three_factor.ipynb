Because the stocks in our portfolio are volatile. When we regress them against the fama-french factors, we may underestimate them. Thus we regress them against the current factors and their lagged versions. 
This files contains the code to create the lagged factors.

```python
import pandas as pd
import numpy as np


fama_french_factor_file = '../data/F-F_Research_Data_Factors_daily.CSV'

fama_french_factor = pd.read_csv(fama_french_factor_file)
# Rename Unamed: 0 to date
fama_french_factor = fama_french_factor.rename(columns={'Unnamed: 0': 'date'})
display(fama_french_factor)
```

```python
# Parse date
fama_french_factor['date'] = pd.to_datetime(fama_french_factor['date'], format='%Y%m%d')
```

```python
# Create lagged factors
fama_french_factor['Mkt-RF_lag1'] = fama_french_factor['Mkt-RF'].shift(1)
fama_french_factor['SMB_lag1'] = fama_french_factor['SMB'].shift(1)
fama_french_factor['HML_lag1'] = fama_french_factor['HML'].shift(1)

# Drop first row
fama_french_factor = fama_french_factor.dropna()
```

```python
display(fama_french_factor)
```

```python
# Save to CSV and Pickle
fama_french_factor.to_csv('../data/fama_french_factor_lagged.csv', index=False)
fama_french_factor.to_pickle('../data/fama_french_factor_lagged.pkl')
```
