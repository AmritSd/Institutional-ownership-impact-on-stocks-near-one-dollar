```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data
price_data = pd.read_pickle('./../data/merged_price_instit_data.pkl')
df = pd.read_pickle('./../data/all_trades_at_5.pkl')

display(price_data)
display(df)
```
